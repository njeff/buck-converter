```python
from scipy import signal
```

```python
# Vref to output
F = 0.5;
Vin = 10;
VM = 5;
H = 1e5;
L = 10e-6;
C = 1e-6;
R = 5;
wn = 1/np.sqrt(L*C)
Q = R*np.sqrt(C/L)

sys = signal.TransferFunction([F*Vin], [1/(wn**2)*VM, 1/(Q*wn)*VM, VM+H*F*Vin])
```

```python
w, mag, phase = signal.bode(sys)
plt.figure()
plt.semilogx(w/(2*np.pi), mag)    # Bode magnitude plot
plt.grid()
plt.figure()
plt.semilogx(w/(2*np.pi), phase)  # Bode phase plot
plt.grid()
plt.show()
```

```python
#Vin to output
F = 0.5;
Vin = 10;
VM = 5;
H = 1e5;
L = 10e-6;
C = 100e-6;
R = 5;
wn = 1/np.sqrt(L*C)
Q = R*np.sqrt(C/L)
D = 0.5;
A = 100;
wp = 314*100; #rad

sys = signal.TransferFunction([D/wp, D], [1/(wn**2 * wp), 1/(wn**2)+1/(Q*wn*wp), 1/(Q*wn) + 1/(wp), 1+A*F])
w, mag, phase = signal.bode(sys)
plt.figure()
plt.semilogx(w/(2*np.pi), mag)    # Bode magnitude plot
plt.grid()
plt.figure()
plt.semilogx(w/(2*np.pi), phase)  # Bode phase plot
plt.grid()
plt.show()
```
