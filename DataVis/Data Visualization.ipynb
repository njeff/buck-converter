```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ideal buck, 1MHz
data = np.loadtxt("../ideal-sim/ideal-buck-1M.txt", skiprows=1, delimiter="\t")

plt.figure(figsize=(10,7))
start=3200
plt.plot(data[start:,0]*1e6, data[start:,1])
plt.grid()

plt.title('1MHz Vout', fontsize=20)
plt.xlabel('Time (us)', fontsize=14)
plt.ylabel('Vout (V)', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('../img/ideal-sim-1MHz-ss.png')
```

```python
# ESR
# Ideal buck, 1MHz
data = np.loadtxt("../ideal-sim/ideal-buck-1M.txt", skiprows=1, delimiter="\t")
data2 = np.loadtxt("../esr-sim/esr-buck.txt", skiprows=1, delimiter="\t")

plt.figure(figsize=(10,7))
stpt = 3200
plt.plot(data[stpt:,0]*1e6, data[stpt:,1], label='Ideal')
plt.plot(data2[int(stpt*1.05):,0]*1e6, data2[int(stpt*1.05):,1], label='w/1 ohm ESR')
plt.grid()
plt.title('Vout: Capacitor ESR Comparison', fontsize=20)
plt.xlabel('Time (us)', fontsize=14)
plt.ylabel('Vout (V)', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend()
plt.savefig('../img/esr_comp.png')
```
